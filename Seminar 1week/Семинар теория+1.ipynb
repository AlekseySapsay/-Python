# Поиск строк с одинаковыми первыми и последними символами
Basic Python Exersises by Google

```python
words= ['a','aaa','aaaaa','abracadabra','kwa','kwakwa']
```

```python
def math_ends(L):
    counter = 0
    for word in L:
        if (len(word) > 1) and (word[0] == word [-1]):
            counter += 1
    return counter            
```

```python
math_ends(words)
```

# B.front_x
Сортировка списка по букве Х. Сначала выводим слова начинающиеся на Х, после- все остальные

```python
words =['min','max','iks','extra','x- files','xyz', 'xapple', 'apple']
```

```python
x_first=[]

for word in words:
    if word[0] == 'x':
        x_first.append(word)
        
x_first
```

```python
x_first = [word for word in words if word[0] =='x']
x_first
```

```python
def front_x(words):
    x_first = [word for word in words if word[0] =='x']
    not_x_first = [word for word in words if word[0] !='x']
    
    return sorted(x_first)+sorted(not_x_first)
```

```python
front_x(words)
```

# C.Sort_last
Сортировка кортежей по последнему символу

```python
tuples=[(1,2), (2,1), (3,5), (10,11), (12,333)]
```

```python
sorted(tuples)
```

```python
def last(elem):
    return elem[-1]
```

```python
sorted(tuples, key = last)
```

```python
sorted(tuples, key= lambda elem: elem[-1])
```
